# Сравнение PySpark и pandas

| Критерий | PySpark DataFrame | pandas DataFrame |
|----------|------------------|------------------|
| **Объём данных** | Работает с очень большими данными (гигабайты и терабайты) распределённо | Подходит только для данных, помещающихся в оперативную память |
| **Обработка** | Параллельная, распределённая обработка (кластер Spark) | Последовательная обработка на одной машине |
| **Скорость** | Высокая при больших объёмах данных | Быстрая при малых объёмах, но медленная при больших |
| **API и синтаксис** | Похож на pandas, но требует явного использования методов Spark (select, filter, orderBy) | Удобный синтаксис, простые операции с колонками |
| **Использование памяти** | Эффективно распределяет нагрузку между узлами | Загружает всё в память — может вызвать переполнение |
| **Установка и запуск** | Требует настройки SparkSession и среды (например, PySpark в Colab) | Достаточно установить pandas (`pip install pandas`) |
| **Сохранение данных** | Использует `df.write.csv()` — может записывать в распределённое хранилище | Использует `df.to_csv()` — сохраняет локально |
| **Применение** | Аналитика больших данных, распределённые вычисления, ETL-процессы | Малые и средние наборы данных, исследовательский анализ |
